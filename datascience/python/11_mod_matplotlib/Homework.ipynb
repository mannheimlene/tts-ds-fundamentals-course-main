# Exercises

Use the data saved at ```./data/company_sales_data.csv``` to use for all exercises.

### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

%matplotlib inline
````

````python
sales = pd.read_csv("data/company_sales_data.csv")
sales
````

````python
x = sales["month_number"]
profit = sales["total_profit"]

plt.plot(x, profit)
plt.title("Total Profit by Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
````

### Exercise 2. 

Get total units of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
units = sales["total_profit"]

plt.plot(x, units, 'o', linestyle=":", color='red', linewidth=3, label='# of Units Sold')
plt.title("Total Units Sold By Month")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(loc='lower right')
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
fig = plt.figure(figsize=(11, 5))
plt.plot(x, sales.facecream, label="facecream")
plt.plot(x, sales.facewash, label="facewash")
plt.plot(x, sales.toothpaste, label="toothpaste")
plt.plot(x, sales.bathingsoap, label="bathing soap")
plt.plot(x, sales.shampoo, label="shampoo")
plt.plot(x, sales.moisturizer, label="moisturizer")
plt.legend(loc="upper left")
plt.title("Units of Product Sold By Month")
plt.xlabel("Month Number")
plt.ylabel("# of Units Sold")
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
plt.scatter(x, sales.toothpaste)
plt.grid(which='both', axis='both', linestyle='-')
plt.title("Units of Toothpaste Sold By Month")
plt.xlabel("Month Number")
plt.ylabel("# of Units Sold")
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
# Exercise 5
````

````python
plt.bar(sales.month_number, sales.facewash, color='blue', label='Facewash', width = 0.25)
plt.bar(sales.month_number + .25, sales.facecream, color='green', label='Facecream', width = 0.25)
plt.legend(loc='upper left')
plt.title("Units of Facewash and Facecream Sold By Month")
plt.xlabel("Month Number")
plt.ylabel("Number Units Sold")
````
